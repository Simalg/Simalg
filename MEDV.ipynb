```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ayarları belirliyoruz
sns.set(style='whitegrid')
%matplotlib inline

```

```python
# Titanic veri setini yüklüyoruz
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)

# Veri setinin ilk 5 satırını görüntülüyoruz
print(df.head())
```

```python
# Eksik değerleri kontrol ediyoruz
print(df.isnull().sum())
```

```python
# Yaş dağılımının KDE ile görselleştirilmesi
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Age'], shade=True)
plt.title('Kernel Density Estimation of Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

```python
# Sınıf ve cinsiyet bazında yaş dağılımının görselleştirilmesi
plt.figure(figsize=(10, 6))
sns.kdeplot(x='Age', hue='Sex', data=df, shade=True)
plt.title('Kernel Density Estimation of Age by Sex')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Sex')
plt.show()
```

```python
# Yaş dağılımının hayatta kalma durumu bazında görselleştirilmesi
plt.figure(figsize=(10, 6))
sns.kdeplot(x='Age', hue='Survived', data=df, shade=True)
plt.title('Kernel Density Estimation of Age by Survival Status')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.show()
```
